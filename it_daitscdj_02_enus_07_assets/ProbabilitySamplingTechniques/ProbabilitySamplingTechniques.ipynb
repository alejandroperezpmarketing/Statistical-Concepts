Stratified Sampling: 
 - Stratified sampling is a type of sampling method in which the total population is divided into smaller groups or strata to complete the sampling process. The strata is formed based on some common characteristics in the population data.

```python
samp_str1 = bestsellers_data.groupby('Genre')\
                            .apply(lambda x: x.sample(8))

samp_str1
```

```python
samp_str1 = bestsellers_data.groupby('Genre', group_keys=False)\
                            .apply(lambda x: x.sample(8))

samp_str1
```

```python
samp_str2 = bestsellers_data.groupby('Genre', group_keys=False)\
                            .apply(lambda x: x.sample(15))

samp_str2
```

```python
def stratified_sampling(df, col, n_samples):
    
    n = min(n_samples, df[col].value_counts().min())
    
    df_groupby = df.groupby(col).apply(lambda x: x.sample(n))
    
    df_groupby.index = df_groupby.index.droplevel(0)
    
    return df_groupby
```

```python
col = ['Genre']
```

```python
n_samples = 4
```

```python
stratified_sample = stratified_sampling(bestsellers_data, col, n_samples)

stratified_sample
```

### Systematic Sampling
 - Systematic sampling is a probability sampling method where researchers select members of the population at a regular interval

Define total number of covid19_constructed_data:

```python
total = len(bestsellers_data)

total
```

Defining Size of Systematic Sample

```python
samples = 25
```

Defining Interval(gap) in order to get required data.

```python
interval = (total // samples)

interval
```

Define systematic sampling function

```python
def systematic_sampling(df, start_index, step):
    
    indexes = np.arange(start_index, len(df), step=step)
    
    systematic_sample = df.iloc[indexes]
    
    return systematic_sample
```

```python
systematic_sample = systematic_sampling(bestsellers_data, 0, interval)

systematic_sample
```

```python
systematic_sample.shape
```

```python
def systematic_sampling_random_start(df, num_samples):
    
    start_index = random.randint(0, len(df)-num_samples+1)
    
    step = (len(df)-start_index) // num_samples
    
    print('Generating', num_samples, 
          'samples with a start index of', start_index,
          'and an interval of', step)
    
    indexes = np.arange(start_index, len(df), step=step)
    
    systematic_sample = df.iloc[indexes]
    
    return systematic_sample
```

Obtain a systematic sample:

```python
systematic_sample = systematic_sampling_random_start(bestsellers_data, 15)

systematic_sample
```
